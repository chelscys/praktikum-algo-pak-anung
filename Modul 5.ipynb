```python
# Chelscy Maria Margaretha
# 065002500018

def konversi_nilai(huruf):
    nilai_huruf = {
        "A": 4.00,
        "A-": 3.75,
        "B+": 3.50,
        "B": 3.00,
        "B-": 2.75,
        "C+": 2.50,
        "C": 2.00,
        "C-": 1.75,
        "D": 1.50,
        "E": 1.25
    }

    huruf = huruf.upper()
    if huruf in nilai_huruf:
        return nilai_huruf[huruf]
    else:
        return 0


total = 0
jumlah = 0

while True:
    huruf = input("Masukkan nilai: ")
    if huruf == "":
        break
    nilai = konversi_nilai(huruf)
    if nilai == 0:
      print("Input tidak valid, coba lagi.")
    else:
      print("Nilai =", nilai)
      total += nilai
      jumlah += 1

if jumlah > 0:
    rata_rata = total / jumlah
    print("Rata - rata nya adalah:", rata_rata)
else:
    print("Tidak ada nilai yang dimasukkan.")
```

```python
# Chelscy Maria Margaretha
# 065002500018

total_harga = 0.0
print("Program Hitung Tiket Kebun Binatang")

while True:
    umur_input = input("Masukkan umur (kosongkan untuk bayar): ")

    if umur_input == "":
        break

    umur = int(umur_input)
    harga_tiket = 0.0

    if umur <= 2:
        print("Gratis")
    elif umur <= 12:
        harga_tiket = 14.00
    elif umur >= 65:
        harga_tiket = 18.00
    else:
        harga_tiket = 23.00

    if harga_tiket > 0:
        print(f"Harga $ {harga_tiket:.2f}")

    total_harga += harga_tiket
    print(f"Running total: {total_harga:.2f}")

print(f"Total Tagihan Anda: ${total_harga:.2f}")
uang_bayar = float(input("Masukkan jumlah uang: "))
kembalian = uang_bayar - total_harga

if kembalian >= 0:
    print(f"Running kembalian: {kembalian:.2f}")
else:
    print(f"Maaf, uang Anda kurang ${-kembalian:.2f}.")
```

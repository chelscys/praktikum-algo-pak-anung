```python
# Chelscy Maria Margaretha
# 065002500018

def konversi_nilai(huruf):
    if huruf == 'A':
        return 4
    elif huruf == 'B':
        return 3
    elif huruf == 'C':
        return 2
    elif huruf == 'D':
        return 1
    elif huruf == 'E':
        return 0
    else:
        print("Huruf tidak valid.")
        return 0

def hitung_rata(total=0, jumlah=0):
    if jumlah == 0:
        return 0
    return total / jumlah

total = 0
jumlah = 0

while True:
    huruf = input("Masukkan nilai: ").upper()
    if huruf == "":
        break
    nilai = konversi_nilai(huruf)

    if nilai != 0:
        print("Nilai =", nilai)
        total += nilai
        jumlah += 1

rata_rata = hitung_rata(total, jumlah)
print("Rata-ratanya adalah:", rata_rata)
```

```python
# Chelscy Maria Margaretha
# 065002500018

def cek_kabisat(tahun):
    if (tahun % 400 == 0) or (tahun % 4 == 0 and tahun % 100 != 0):
        return True
    else:
        return False


def jumlah_hari(bulan, tahun):
    if bulan == 2:
        if cek_kabisat(tahun):
            return 29
        else:
            return 28
    elif bulan in [4, 6, 9, 11]:
        return 30
    else:
        return 31


print("Program ini akan menentukan jumlah hari dalam bulan tertentu")
print("Masukkan 0 untuk menghentikan program")

while True:
    bulan = int(input("Masukkan bulan (1-12): "))
    if bulan == 0:
        print("Terima kasih sudah menggunakan program saya, sampai berjumpa lagi.")
        break

    if bulan == 2:
        tahun = int(input("Masukkan tahun (e.g., 2021): "))
        hari = jumlah_hari(bulan, tahun)
    else:
        hari = jumlah_hari(bulan, 0)

    print(f"{hari} hari dalam sebulan.")
```
